```python
# Part 1)
    
file1 = open('sars-cov2_spike_protein_mRNA.fasta', 'r') 

# Part 2)
# Function to convert a FASTA file into a DNA sequence string object.

data = file1.readlines()[1:]
str1 = ''.join(data)
```

```python
# Codon to amino acid reference table 

codon_dict = { # https://www.geeksforgeeks.org/dna-protein-python-3/
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K', 
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',                  
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L', 
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q', 
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V', 
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G', 
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L', 
    'TAC':'Y', 'TAT':'Y', 'TAA':'*', 'TAG':'*', 
    'TGC':'C', 'TGT':'C', 'TGA':'*', 'TGG':'W', 
}
```

```python
# Part 3)
# 3.1 generate list with sense and anti-sense sequence strings

complementer_dict = {'A':'T', 'T':'A', 'G':'C', 'C':'G'}

sense_seq_str = str1.replace('\n','')
sense_seq_list = list(sense_seq_str)

anti_seq_list = [complementer_dict[base] for base in sense_seq_list]
anti_seq_str = ''.join(anti_seq_list)

sense_n_anti_list = [sense_seq_str, anti_seq_str]
```

```python
# 3.2 split base sequence strings into list of codons 
# generate list of 6 lists of codons ~3x for both sense and anti strings in sense_n_anti_list = 6x lists of codon lists

n = 3
codonlists = []

for i in range(2): 
    currentstrand = sense_n_anti_list[i]
    for k in range(3):
        codonlists.append([currentstrand[j:j+n] for j in range(k, len(sense_seq_str)-k, n)])
    
print(codonlists)
```

```python
# Function to translate codon lists into peptide sequences

list_of_peptides = []

for currentcodon_list in codonlists:
    list_of_aas =[]
    for trimer in currentcodon_list:
        if len(trimer) == 3:
            aa = codon_dict[trimer]
        else: 
            continue
        list_of_aas.append(aa)      
    list_of_peptides.append(list_of_aas)

print(list_of_peptides)
```
